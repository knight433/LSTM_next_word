```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.models import load_model
```

```python
# Tokenizer initialization and fitting
tokenizer = Tokenizer()
tokenizer.fit_on_texts([faqs])

# Creating input sequences
input_sequences = []
for sentence in faqs.split('\n'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1, len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])

# Padding sequences
max_len = max([len(x) for x in input_sequences])
padded_input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')

# Creating training data
X = padded_input_sequences[:, :-1]
y = padded_input_sequences[:, -1]

# One-hot encoding the labels
num_classes = len(tokenizer.word_index) + 1
y = to_categorical(y, num_classes=num_classes)

vocab_size = num_classes  # Number of unique tokens + 1
embedding_dim = 100
input_length = max_len - 1  # Because X is padded_input_sequences[:, :-1]
```

```python
import json

# Save tokenizer configuration and word index
tokenizer_config = tokenizer.to_json()
with open('tokenizer_config.json', 'w') as f:
    json.dump(tokenizer_config, f)
    
word_index = tokenizer.word_index
with open('word_index.json', 'w') as f:
    json.dump(word_index, f)
```

```python
from tensorflow.keras.models import load_model
model = load_model('my_model.h5')
model.fit(X, y, epochs=50)

# Save the model again
model.save('my_model_continued.h5')
```

```python

model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=input_length))
model.add(LSTM(150, return_sequences=True))
model.add(LSTM(150))
model.add(Dense(vocab_size, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```
